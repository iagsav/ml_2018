```python
import pandas as pd
from scipy import stats
import seaborn as sns
df = pd.read_csv('C:\\auto.csv', encoding = 'latin1')
df
```

Удалите столбцы, ценность которых для оценки стоимости машины низка. Удалите повторяющиеся строки, строки содержащие пропуски в данных. Выведите размер набора данных до и после удаления.

```python
print(df.shape)
print("размер данных до удаления")
del df['dateCrawled']
del df['seller']
del df['offerType']
del df['abtest']
del df['dateCreated']
del df['nrOfPictures']
del df['lastSeen']
df = df.dropna()
df = df.drop_duplicates()
df = df.drop(df[df.monthOfRegistration == 0].index)
df = df.drop(df[df.powerPS == 0].index)
df = df.reset_index()
del df['index']
print(df.shape)
print("размер данных после удаления")
```

Удалите данные в строках, выходящие за некоторые пределы (год регистрации, цена, мощность двигателя). Выведите размер набора данных до и после удаления. Для оценки диапазонов значений признаков используйте квантили.

```python
qu = df.quantile([.15,.5,.85])
df = df.drop(df[df.price < qu['price'][0.15]].index)
df = df.drop(df[df.price > qu['price'][0.85]].index)
df = df.drop(df[df.yearOfRegistration < qu['yearOfRegistration'][0.15]].index)
df = df.drop(df[df.yearOfRegistration > qu['yearOfRegistration'][0.85]].index)
df = df.drop(df[df.powerPS < qu['powerPS'][0.15]].index)
df = df.drop(df[df.powerPS > qu['powerPS'][0.85]].index)
df = df.reset_index()
del df['index']
print(df.shape)
```

Определите коэффициент корреляции Пирсона и Спирмена между стоимостью автомобиля и типом кузова. Оцените значение p-value. Постройте плотность распределения признаков.

```python
stats.spearmanr(df.price,df.vehicleType)
```

```python
arr = {'limousine':1, 'kombi':2, 'bus':3, 'kleinwagen':4, 'cabrio':5, 'suv':6, 'coupe':7, 'andere':8}
df['pearson'] = df['vehicleType']
for i in range(len(df['vehicleType'])):
    df['pearson'].values[i] = arr[df['vehicleType'].values[i]]

stats.pearsonr(df.price,df.pearson)
```

```python
s = pd.Series(df.pearson)
s.plot(kind = "kde")
```
