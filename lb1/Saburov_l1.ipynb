### 1. Постройте графики количества осадков по годам в разных штатах. Сделайте выводы по построенным графикам.

```python
arrSD1 = tuple(df['SUBDIVISION'].unique())
for i in range(0,len(arrSD1)):
    df1 = df[['SUBDIVISION','YEAR','ANNUAL']]
    df1 = df1[df1['SUBDIVISION'] == arrSD1[i]]  
    df1.plot(x='YEAR', y='ANNUAL', title=str(i+1) + ") " + str(arrSD1[i]))
    plt.show()
```

### Выводы: По построенным графикам видно, что в разных штатах в разные года наблюдений количество осадков может координально отличаться. С помошью графиков можно получить представление и понимание о том, как изменялось количество осадков в различных штатах с течение времени.

### 2. Постройте графики количества осадков по месяцам по всем годам кумулятивно. Сделайте выводы по построенным графикам.

```python
for i in range(2,14):
    df2 = df.iloc[:,[1,i]]
    res2 = 0
    lst21 = []
    lst22 = []
    for j in range(df2[df2.columns[0]].min(),(df2[df2.columns[0]].max())+1):
        df22 = df2.copy()
        df22 = df22[df22['YEAR'] == j]
        res2 += df22[df22.columns[1]].sum()
        lst21.append(j)
        lst22.append(res2)
    tup21 = ('YEAR',lst21)
    tup22 = ('OVERALL',lst22)
    tmp2 = dict([tup21, tup22])
    df2tmp = pd.DataFrame(tmp2)
    df2tmp.plot(x='YEAR', y='OVERALL', title = str(i-1) + ") " + str(df2.columns[1]))
    plt.show()
```

### Выводы: По данным графикам можно приблизительно определить, что в каждом месяце, кривая количества осадков за период наблюдений изменяется более-менее линейно, следовательно, каждый год, во всех штатах суммарно, выпадает относительно равное количество осадков.
