## 2 задание

```python
from scipy import stats
```

```python
na = nba["Age"].corr(nba["G"])
```

```python
na
```

```python
n = stats.spearmanr(nba["Age"],nba["G"], nan_policy="omit")
```

```python
print(n[0])
```

```python
if n[0] > na:
    print ("Корреляция Спирмена больше Пирсона")
elif n[0] == na:
    print ("Корреляция Спирмена ровна Пирсону")
else:
    print ("Корреляция Спирмена меньше Пирсона")
```
