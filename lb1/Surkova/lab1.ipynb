```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as mtp
import scipy as sc
mtp.rcParams.update({'font.size':18,'figure.figsize':[30,25]})    #устанавливаем размер шрифта и размер графиков
```

```python
df = pd.read_csv('D:\\rainfall_in_india_1901-2015.csv')
df
```

### 1. Постройте графики количества осадков по годам в разных штатах. Сделайте выводы по построенным графикам.

```python
fig, axes = mtp.subplots(6,2)
i=0
j=0
count=0
for lbl, grp in df.groupby(['SUBDIVISION']):
    grp.plot(ax=axes[i][j], kind='line', x='YEAR', y='ANNUAL',label = lbl)
    count+=1
    if count%3==0:
        if j==1:
            i+=1
            j=0
        else:
            j+=1
```

##### Как видно из графиков, наименьшее количество осадков бывает в штатах East Rajastan, Saurashtra & Kutch, West Rajastan. Это может зависеть от расположения штатов, так как в таблице они представлены в алфавитном порядке, и не отражают географического расположения. Наибольшее количество осадков - в штатах Arunachal Pradesh, Coastal Karnataka.

### 2. Постройте график количества осадков по месяцам по всем годам кумулятивно. Сделайте выводы по построенным графикам.

```python
table=pd.DataFrame({'month':['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec'],
                    'rains':[df['JAN'].sum(),df['FEB'].sum(),df['MAR'].sum(),df['APR'].sum(),df['MAY'].sum(),df['JUN'].sum(),
                             df['JUL'].sum(),df['AUG'].sum(),df['SEP'].sum(),df['OCT'].sum(),df['NOV'].sum(),df['DEC'].sum()]})
table
```

```python
table.index=['january','february','march','april','may','june','jule','august','september','october','november','december']
table.plot(kind='bar',legend=False)
```

##### Из гистограммы видно, что наибольшее количество осадков бывает летом - в июле и августе, наименьшее - в зимние месяцы.

### 4. Определите штаты, для которых количество осадков минимально и максимально.

```python
mnmx = df.groupby(["SUBDIVISION"])["ANNUAL"].agg("sum")
print("Минимальное количество осадков - {} - {}".format(mnmx.idxmin(), mnmx.min()))
print("Максимальное количество осадков - {} - {}".format(mnmx.idxmax(), mnmx.max()))
```
